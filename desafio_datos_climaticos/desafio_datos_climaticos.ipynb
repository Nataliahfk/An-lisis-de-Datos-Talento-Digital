```python
import pandas as pd
import numpy as np
from datetime import date
```

## 1. Extracción de Datos:

```python
# 1. Extracción de Datos: Leer el archivo CSV
df_clima = pd.read_csv('datos_climaticos.csv', sep=',', encoding='utf-8')
df_clima
```

```python
df_clima.info()
```

```python
'''2. Extraer columnas temperatura_celsius y humedad_relativa y conviértelas en un único arreglo 
(array) de NumPy ''' 

datos_numericos = np.array(df_clima[['temperatura_celsius', 'humedad_relativa']])
datos_numericos
```

```python
type(datos_numericos)
```

```python
datos_numericos.shape
```

## 2. Manipulación y Análisis con NumPy

#### - Usando el arreglo datos_numericos:

##### a. Selección: 
Selecciona e imprime únicamente la primera hora de mediciones(las primeras 12 filas) de ambas variables.

```python
# primera hora de mediciones (las primeras 12 filas) de ambas variables.

primera_hora = datos_numericos[:12,:]
print('Primeras 12 mediciones (Temperatura y Humedad): \n', primera_hora)
```

```python
primera_hora.shape
```

##### b. Filtrado Condicional: 
Encuentra y muestra todas las temperaturas que fueron superiores a 25.0°C.

```python
temperaturas = datos_numericos[:,0] # trae todas las filas de la columna 0
temperaturas
```

```python
temperatura_superior = temperaturas[temperaturas > 25.0]
print('\nTemperaturas superiores a 25.0°C:\n', temperatura_superior)
```

```python
temperatura_superior.shape
```

##### c. Operaciones: 
Convierte todos los valores de temperatura de Celsius a Fahrenheit. La fórmula es: F=C×9/5 + 32. Almacena el resultado en un nuevo vector llamado temperatura_fahrenheit

```python
# Fórmula: F = C * (9/5) + 32  // Hay un error en la guíaen la fórmula
temperatura_fahrenheit = (temperaturas * 9/5) + 32
print('Temperaturas Fahrenheit:\n', temperatura_fahrenheit)
```

##### d. Creación y Redimensión: 
Crea una matriz de 2x2 que contenga la primera y la última medición de temperatura y humedad del día.

```python
# Matriz 2x2
# Extraer la primera y la 'ultima fila de la matriz datos_numericos
primera_medicion = datos_numericos[0]
ultima_medicion = datos_numericos[-1]


matriz_primera_ultima = np.array([primera_medicion,ultima_medicion])
print('\nMatriz 2x2 (Inicio y Fin del día):\n', matriz_primera_ultima)
```

### 3. Escritura de Datos:

○ Guarda el nuevo vector temperatura_fahrenheit que creaste en un nuevo archivo llamado temperaturas_fahrenheit.csv.

```python
hoy = date.today()
print(hoy)
```

```python
# Nombre archivo
nombre_archivo = (f'temperaturas_fahrenheit.csv')

# Crear DataFrame para guardar vector temperatura_fahrenheit para usar .to_csv
df_temperatura_fahrenheit = pd.DataFrame(temperatura_fahrenheit, columns=['temperatura_fahrenheit'])

df_temperatura_fahrenheit.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')

print(f'Archivo generado con éxito: {nombre_archivo}_{hoy}')
```

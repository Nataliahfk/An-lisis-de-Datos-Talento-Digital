Para eso, deberás simular lecturas de temperatura, humedad y luminosidad en distintas condiciones (rango fijo, aleatoriedad, calibración de referencia).

```python
import numpy as np
```

```python
# 1. Simular 10 lecturas de temperatura entre 15°C y 30°C con linspace
temperaturas = np.linspace(15,30,10)
print('Lecturas simuladas de temperiaturas (C°): \n', temperaturas)
```

```python
# 2. Crear una matriz identidad para validar referencia (ideal de calibración)
calibracion = np.identity(3) # 3 caracteristicas que almacena
print("\nMatriz identidad de referencia para sensores:\n", calibracion)
```

```python
# 3. Generar 5 lecturas aleatorias de humedad entre 40% y 80%
np.random.seed(0)
humedad = np.random.randint(40, 81, size=5)
print("\nLecturas aleatorias de humedad (%):\n", humedad)
```

```python
# 4. Simular 3 series de luminosidad entre 200 y 800 lúmenes.
# Los sensores trabajaron de manera sucesiva.
np.random.seed(0)
luminosidad = np.random.randint(200,801, size=(3,5))
print("\nLecturas de luminosidad en 3 sensores (lúmenes):\n", luminosidad)
```

```python
# 5. Reestructurar datos de luminosidad para análisis por minuto
# flatten para que cree un vector 
luminosidad_lineal = luminosidad.flatten()
print("\nLecturas de luminosidad aplanadas (por minuto):\n", luminosidad_lineal)
```

```python
# 6. Preparar arreglo para insertar nuevas mediciones
estructura = np.zeros((3, 5))
estructura
```

```python
# ejemplo: El slicing dentro del selector [] permite seleccionar filas o columnas para aplicar el random
ejemplo = estructura
ejemplo [1 , :] = np.random.randint(100, 500, size=5) 
print("\nEstructura lista para nuevas mediciones (datos aleatorios insertados):\n", ejemplo)
```

```python
ejemplo [2,:]= np.random.randint(100, 500, size= (5,)) 
print("\nEstructura lista para nuevas mediciones (datos aleatorios insertados):\n", ejemplo)
```

```python
# 6. Preparar arreglo para insertar nuevas mediciones
# [: , :] es para que el llenado de la matriz de ceros sea completa todas las filas y todas las columnas 
estructura[: , :] = np.random.randint(100, 500, size=(3, 5)) 
print("\nEstructura lista para nuevas mediciones (datos aleatorios insertados):\n", estructura)
```

## Vector con distribución de puntos

La función **np.linspace()** permite crear un vector de valores numéricos igualmente espaciados entre dos extremos, ideal para representar escalas, tiempo, intervalos de medición o datos normalizados. A diferencia de escribir listas manuales o usar np.arange(), esta función garantiza que se generen exactamente n puntos entre los extremos especificados, incluyendo
ambos límites.

### ✅ np.linspace()

👉 Sirve para crear números entre un inicio y un fin, indicando cuántos valores quieres.

**np.linspace(inicio, fin, cantidad)**

```python
import numpy as np

x = np.linspace(0, 1, 5) # considera inicio y final
print(x)
```

👉 Aquí le dijiste:

empieza en 0

termina en 1

quiero 5 valores

NumPy calcula solo el paso.

👉 Muy importante

✔ linspace SIEMPRE incluye el valor final (por defecto).

### ✅ np.arange()

👉 Sirve para crear números desde un inicio hasta un fin usando un paso fijo.

👉 Forma:

**np.arange(inicio, fin, paso)**

```python
x = np.arange(0, 1, 0.25) # considera inicio y no el final (1)
print(x)
```

👉 Aquí le dijiste:

empieza en 0

llega hasta antes de 1

avanza de 0.25

👉 Muy importante

✔ arange NO asegura incluir el final.

- 🧩 En tu nivel (y para evitar errores)

👉 Usa:

✅ linspace cuando te pidan:
“genera N valores entre A y B”

✅ arange cuando te pidan:
“genera valores cada cierto paso”

- 👀 Resumen muy corto

👉 linspace → cantidad
👉 arange → paso

🧩 Conectándolo con tu ejercicio anterior

Tú tenías algo como:

tiempos de lectura cada 0.25

👉 Ahí está bien:

```python
np.arange(0, 4, 0.25)
```

Pero si te piden:

genera exactamente 16 tiempos entre 0 y 4

👉 ahí conviene:

```python
# con 16 valores me da un intervalo 0,266666
np.linspace(0,4,16)
```

Tiene:

👉 16 valores
👉 empieza en 0
👉 termina en 4

Eso calza perfecto con: **linspace(inicio=0, fin=4, cantidad=16)**

📏 Cuál es el paso real

Cuando usas linspace(0, 4, 16) el paso no es 0.25.

Es:

(4−0)/(16−1)=4/15=0.2666666666...


⚠️ Ojo con un error muy común

Muchos estudiantes creen:

16 valores entre 0 y 4 → paso = 4/16 = 0.25

❌ Eso sería correcto solo si fuera arange.

Pero en linspace:

👉 el intervalo se reparte en 15 espacios, no en 16.

🔎 Si tú querías paso 0.25 entre 0 y 4

Entonces no debes usar linspace así, sino:

```python
np.arange(0, 4.01, 0.25)
```

(o bien)

```python
# para que el intervalo sea de 0.25 tengo que poner 17 valores
np.linspace(0, 4, 17)
```

Porque:

👉 de 0 a 4 cada 0.25 hay 17 valores.

Características:

- np.linspace() : - Inclusión del límite final / Sí, incluye el valor final

                  - Precisión de espaciado / Preciso (define cantidad de puntos)

                  - Parámetro principal / Número de puntos


- np.arange() :   - Inclusión del límite final / No, excluye el valor final

                  - Precisión de espaciado / A veces impreciso por redondeo
                  
                  - Parámetro principal / Paso o incremento

```python
# otro ejemplo partiendo de 2
ejemplo_2 = np.arange(2,4.1,0.25) # para considerar del 2 al 4 / no es lo ideal
print(ejemplo_2)
```

```python
# para que haga la medicion desde el 2 al 4, sin considerarlo, saltando de 0.25, eso son 8 saltos, pk no llega hasta el 4
ejemplo_2 = np.arange(2,4,0.25) 
print(ejemplo_2)
```

```python
# para que haga la medicion desde el 2 al 4 considerarlo, y le pido 9 elementos para que considere el 4
ejemplo_2 = np.linspace(2,4,9) 
print(ejemplo_2)
```

```python
ejemplo_3 = np.arange(0,0.7,0.3)
print(ejemplo_3)
```

```python
ejemplo_3 = np.linspace(0,0.6,3)
print(ejemplo_3)
```

### Ejemplo de linspace()

```python
# A continuación se presenta un ejemplo de código en el que se genera un vector de 5 valores equiespaciados entre 0 y 1:
npvector = np.linspace(0,1,5)
print(npvector)
```

Este fragmento de código genera un vector de 5 valores distribuidos de forma equidistante entre 0 y 1. Es
especialmente útil cuando se requiere un número exacto de puntos dentro de un intervalo definido.

```python
ejemplo4 = np.arange(0,3,0.33333)
print(ejemplo4)
```

```python
ejemplo4 = np.linspace(0,3,10)
print(ejemplo4)
```

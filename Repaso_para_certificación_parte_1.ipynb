## Operadores aritméticos
1) , +  suma
2) , -  resta
3) , *  multiplicación
4) /  división (resultado siempre float)
5) // división entera (sin decimales)
6) %  módulo (residuo)
7) ** potencia

```python
# Notas de un estudiante
nota1 = 4.5
nota2 = 3.8
nota3 = 4.2

# Cálculos
suma_notas = nota1 + nota2 + nota3
promedio = suma_notas / 3
promedio_entero = suma_notas // 3  # División entera
resto = suma_notas % 3              # Residuo

print(f"Suma: {suma_notas}")
print(f"Promedio: {promedio:.2f}")
print(f"Promedio sin decimales: {promedio_entero}")
print(f"Residuo de la división: {resto:.2f}")

# Potencia: ¿cuántos puntos necesita para llegar a 7?
faltante = 7 - promedio
print(f"Le falta: {faltante:.2f} puntos para la nota máxima")
```

## Operadores de comparación (resultado: True o False)
1) ==  igual que
2) !=  diferente que
3) , >   mayor que
4) , <   menor que
5) , >=  mayor o igual que
6) , <=  menor o igual que

```python
nota_final = 4.2
nota_minima_aprobacion = 4.0

# Comparaciones
es_aprobado = nota_final >= nota_minima_aprobacion
es_excelente = nota_final >= 6.0
es_reprobado = nota_final < 4.0
misma_nota = nota_final == 4.0

print(f"¿Aprueba? {es_aprobado}")
print(f"¿Es excelente? {es_excelente}")
print(f"¿Reprueba? {es_reprobado}")
print(f"¿Nota exactamente 4.0? {misma_nota}")

# Operadores lógicos: and (y), or (o), not (no)
edad = 15
tiene_permiso = True

puede_entrar = (edad >= 14) and tiene_permiso
print(f"\n¿Puede entrar al campamento? {puede_entrar}")
```

# Sentencias repetitivas y estructuras de datos

```python
# Definir la función promedio
def promedio(notas):
    return sum(notas) / len(notas)

# Diccionario de estudiantes
estudiantes = {
    "Ana": [5.2, 6.0, 4.5, 3.8],
    "Luis": [4.0, 3.5, 4.2, 5.0],
    "Carla": [6.5, 7.0, 6.8, 6.2]
}

```

```python
# Ciclo for para calcular y mostrar promedios
print("=== PROMEDIOS POR ESTUDIANTE ===\n")
for nombre, notas in estudiantes.items():
    prom = promedio(notas)
    print(f"{nombre}: {prom:.2f}")

```

```python
# Calcular promedio del curso
print("\n=== PROMEDIO DEL CURSO ===")
promedios = []
for nombre, notas in estudiantes.items():
    promedios.append(promedio(notas))

prom_curso = sum(promedios) / len(promedios)
print(f"Promedio general: {prom_curso:.2f}")
```

# Iteraciones while

🔐 Sistema de inicio de sesión (Intentos hasta acertar)

```python
intentos = 0
contrasena_correcta = "1234"

print("🔒 Sistema de seguridad")

while intentos < 3:
    contrasena = input("Ingrese su contraseña: ")

    if contrasena == contrasena_correcta:
        print("✅ Acceso concedido")
        break  # Sale del bucle
    else:
        intentos = intentos + 1
        print(f"❌ Contraseña incorrecta. Intentos restantes: {3 - intentos}")

if intentos == 3:
    print("🚫 Cuenta bloqueada por demasiados intentos")
```

💰 Ahorro para una meta (Repetir hasta alcanzar objetivo)

```python
meta_ahorro = 100000  # $100,000
ahorrado = 0

print(f"🎯 Meta: ${meta_ahorro}")

while ahorrado < meta_ahorro:
    deposito = float(input("¿Cuánto deseas depositar hoy? $"))
    ahorrado = ahorrado + deposito
    print(f"💰 Llevas ahorrado: ${ahorrado}")
    print(f"📊 Te faltan: ${meta_ahorro - ahorrado}")
    print("-" * 30)

print(f"🎉 ¡Felicidades! Alcanzaste tu meta de ${ahorrado}")
```

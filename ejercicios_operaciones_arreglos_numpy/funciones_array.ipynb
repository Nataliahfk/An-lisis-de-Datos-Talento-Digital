# Aplicando funciones a un arreglo.

¿Cómo puedes calcular transformaciones y estadísticas sobre todos los elementos de un arreglo de forma rápida?

Una de las mayores fortalezas de NumPy es la posibilidad de aplicar funciones matemáticas y estadísticas irectamente sobre arreglos, sin tener que escribir bucles. Esto permite procesar grandes volúmenes de datos con muy pocas líneas de código y con un rendimiento muy superior al de las listas tradicionales de Python.
Estas funciones están optimizadas en C y se ejecutan de forma vectorizada, es decir, se aplican automáticamente a todos los elementos del arreglo.

## Funciones matemáticas elementales

NumPy incluye funciones básicas como:

```python
import numpy as np
```

```python
datos = np.array([1, 4, 9, 16])

print(np.sqrt(datos)) # Raíz cuadrada: [1. 2. 3. 4.]
print(np.log(datos)) # Logaritmo natural
print(np.exp(datos)) # Exponencial e^x
print(np.sin(datos)) # Seno de cada elemento (en radianes no grados)
print(np.cos(datos)) # Coseno de cada elemento (en radianes no grados)
```

Cada una de estas funciones transforma todos los elementos del arreglo de forma simultánea y sin bucles.

## Aplicar funciones estadísticas

Estas funciones son esenciales para el análisis exploratorio de datos.

```python
valores = np.array([10, 20, 30, 40])

print(np.mean(valores)) # Promedio: 25.0
print(np.std(valores)) # Desviación estándar: 11.18
print(np.min(valores)) # Valor mínimo: 10
print(np.max(valores)) # Valor máximo: 40
print(np.sum(valores)) # Suma total: 100
```

## Aplicar funciones sobre matrices

Cuando se trabaja con matrices (2D), NumPy permite aplicar funciones a lo largo de un eje:

```python
matriz = np.array([[1, 2, 3],[4, 5, 6]])

print(matriz)

print(np.sum(matriz, axis=0))# Suma por columna: [1+4= 5  2+5= 7  3+6= 9 ] ---> [5 7 9]
print(np.sum(matriz, axis=1))# Suma por fila: [1+2+3 = 6  4+5+6 = 15 ]---> [6 15] 
```

Esto es especialmente útil en análisis por variable (columnas) o por caso (filas).

## Funciones booleanas y de comparación

NumPy permite generar arreglos booleanos que indican si se cumple una condición:

```python
datos = np.array([3, 7, 2, 9])

print(datos > 5)
# [False True False True]
print(np.any(datos > 5))
# True: al menos un valor cumple
print(np.all(datos > 1))
# True: todos los valores > 1
print(np.any(datos <2))
print(np.all(datos > 4))

```

También es posible contar cuántos cumplen la condición:

```python
print(np.sum(datos > 5))
# Resultado: 2
print(np.sum(datos < 10))
```

## Funciones universales (ufuncs)

Las funciones matemáticas y estadísticas de NumPy son conocidas como ufuncs o funciones universales. Estas funciones están implementadas en C para operar con alta velocidad sobre cada elemento del arreglo.

Características de las ufuncs:

• Se ejecutan sobre arreglos de cualquier forma o tamaño.

• Admiten operaciones entre arreglos o entre un arreglo y un escalar.

• Se pueden combinar para construir expresiones complejas sin sacrificar velocidad.

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(np.add(x, y)) # sumar vectores
# [5 7 9]
print(np.multiply(x, y)) # multiplicar vectores
# [4 10 18]
```

Las ufuncs (universal functions) son funciones que:

👉 trabajan elemento por elemento sobre un arreglo NumPy.

## Composición de funciones

NumPy permite combinar funciones en una sola línea para crear pipelines simples:

En programación y ciencia de datos, el término “pipeline” se refiere a un flujo o secuencia de pasos que se ejecutan de manera organizada y automática, donde la salida de un paso se convierte en la entrada del siguiente. Es como una línea de producción, pero de datos o procesos.

Un pipeline organiza las etapas de transformación de datos y entrenamiento de modelos.

Lectura de datos → Limpieza de datos → Transformación → Entrenamiento del modelo → Predicción

💡 Resumen fácil:
Un pipeline es como una cinta transportadora de pasos automáticos: haces algo en el primer paso, lo que resulta pasa al siguiente, y así hasta terminar todo el proceso.

### Ejemplo: Normalización de datos

Supongamos que recibimos una matriz con mediciones de sensores ambientales y queremos calcular:

• El promedio de cada variable.

• La desviación estándar.

• Un vector con valores normalizados.

```python
''' datos es una matriz de 3×3 (3 filas, 3 columnas).

Cada fila representa una medición de sensores ambientales.

Cada columna representa una variable distinta (por ejemplo, temperatura, humedad, presión).
'''

datos = np.array([
                 [22.5, 55.1, 83.2],
                 [21.9, 53.2, 80.5],
                 [23.0, 54.8, 81.0]
                 ])

media = np.mean(datos, axis=0) # calcula el promedio por columna ej:temperatura, humedad, presión
desviacion = np.std(datos, axis=0) # calcula cuánto se desvían los datos respecto a su media.
normalizado = (datos - media) / desviacion 
# x (datos)= valor original - μ (media)= media de la columna / σ (desviación) = desviación estándar de la columna

print("\nMedia:\n", media)
print("\nDesviacion:\n", desviacion)
print("\nNormalizado:\n", normalizado)
```

Cada fila corresponde a la misma fila de datos original, pero ahora los valores son comparables entre columnas, sin importar la escala original.

Por ejemplo:

El valor 22.5 en la primera columna se convirtió en 0.0741

El valor 55.1 en la segunda columna se convirtió en 0.8792

Esto permite que algoritmos de machine learning no se vean sesgados por la magnitud de las variables.

Lo que estás viendo no es un rango 0 a 1, y por eso hay valores mayores a 1 o menores a -1. Esto se debe a que este ejemplo usa normalización Z o estandarización, no normalización min-max. Te explico la diferencia:

### 2️⃣ Media y desviación con axis

np.mean(datos, axis=0) → calcula la media de cada columna.

np.std(datos, axis=0) → calcula la desviación estándar de cada columna.

Esto es lo que usualmente quieres para estandarización, porque cada característica (columna) puede tener unidades y rangos distintos.

- Si no pones axis:

np.mean(datos)  # media de todos los elementos de la matriz
np.std(datos)   # desviación estándar de todos los elementos


Esto mezcla todas las columnas, lo que NO es correcto si quieres escalar cada columna por separado.

💡 Resumen:

axis=0 → escalar por columna (lo usual en ciencia de datos).

axis=1 → escalar por fila (menos común).

Sin axis → calcula sobre toda la matriz, no por características.

### 1️⃣ Normalización Z (o estandarización) ✅

x = valor original

μ = media de la columna

σ = desviación estándar de la columna

Resultado: los valores están centrados en 0 con desviación 1.

Propiedades:

No están limitados entre 0 y 1.

Pueden ser mayores que 1 o menores que -1 si los valores están lejos de la media.

Útil para machine learning porque todas las variables tienen la misma escala relativa.

En tu ejemplo, los valores como 1.3926 o -1.2601 son perfectamente normales con Z-score.

💡 En resumen:

Tu ejemplo no da 0–1 porque usa Z-score, no Min-Max.

Valores mayores a 1 o menores que -1 son normales en la estandarización.

Si necesitas valores estrictamente entre 0 y 1, debes usar Min-Max scaling.

No, normalización y estandarización no son lo mismo, aunque ambos son métodos para escalar datos. Te explico paso a paso con claridad:

### 1️⃣ Normalización (Min-Max Scaling)

Objetivo: Escalar los datos a un rango específico, usualmente [0, 1] o [-1, 1].

Formula :

Xnorm =  X - Xmin / Xmax - Xmin

- Ejemplo:

Datos: [10, 20, 30]

Mínimo = 10, Máximo = 30

Normalizados a [0,1]:

10 → 10 - 10 / 30-10 = 0/20 = 0

20 → 20 - 10 / 30-10 = 10/20 = 0.5

30 → 30-10 / 30-10 = 20/20 = 1

Uso típico: cuando quieres que todos los valores estén dentro de un rango fijo, útil para algoritmos como redes neuronales o KNN.


### 2️⃣ Estandarización (Z-score Scaling)

Objetivo: Transformar los datos para que tengan media 0 y desviación estándar 1.

Fórmula:

Xest = X-μ / σ

donde 
𝜇 = media, 

σ = desviación estándar.

- Ejemplo:

Datos: [10, 20, 30]

Media = 20, Desviación estándar ≈ 8.16

- Estandarizados:

10 → (10-20)/8.16 ≈ -1.22

20 → (20-20)/8.16 = 0

30 → (30-20)/8.16 ≈ 1.22

Uso típico: cuando los algoritmos asumen distribución normal de los datos (como regresión lineal, SVM, PCA).

| Método          | Rango             | Media/Desviación                   | Uso típico                       |
| --------------- | ----------------- | ---------------------------------- | -------------------------------- |
| Normalización   | 0 a 1 (o -1 a 1)  | No cambia la forma de distribución | Redes neuronales, KNN            |
| Estandarización | No hay rango fijo | Media=0, Desviación=1              | Algoritmos que asumen normalidad |

# Referencia y copia de arreglos en NumPy.

¿Qué sucede cuando asignas un arreglo a otra variable? ¿Es una copia o una referencia?

En Python, las variables funcionan como referencias a objetos en memoria. Cuando trabajamos con arreglos en NumPy, es
fundamental entender esta diferencia, ya que una asignación puede no crear una copia nueva, sino simplemente otro nombre que apunta al mismo arreglo original. Esto puede llevar a errores sutiles y difíciles de detectar si se modifican datos sin
intención.

```python
import numpy as np
```

## Asignación directa: referencia al mismo arreglo

```python
a = np.array([1,2,3])
b = a # No se crea una copia, b apunta al mismo arreglo
b[0] = 100
print("a:", a) # Resultado: [100 2 3]
print("b:", b) # Resultado: [100 2 3]
```

```python
c = np.array([10,20,30])
d = np.arange(10,20,4)

print(c)
print(d)
```

¿Qué ocurrió? 

Al modificar b, también se modificó a, porque ambas
variables referencian exactamente el mismo objeto en memoria. Esto
se denomina aliasing.

### ¿Cómo verificar si dos arreglos comparten memoria?

Podemos usar np.shares_memory() para confirmar si dos arreglos apuntan al mismo bloque de datos:

```python
np.shares_memory(a,b)
```

```python
np.shares_memory(c,d)
```

# Crear una copia real del arreglo 

Si deseamos trabajar con una copia completamente independiente, debemos usar el método .copy():

```python
print(a)
```

```python
copia = a.copy()
print(copia)
```

```python
copia[0] = 999
print('a:', a)
print('copia:', copia)

# Ahora, copia es una nueva estructura de datos, y los cambios en ella no afectan a a.
```

# Copias en matrices: el caso con slicing

```python
matriz = np.array([[1, 2], [3, 4]])
submatriz = matriz[:, 0] # Selecciona la primera columna
print("submatriz:\n", submatriz)
```

```python
submatriz[0] = 100
print("matriz:\n", matriz)
print("submatriz:\n", submatriz)
```

Esto sucede porque submatriz no es una copia independiente, sino una vista (view) sobre la matriz original.

# Cómo hacer una copia segura de una subestructura

```python
# al hacer el copy no afectaria las modificaciones de submatriz sin afectar la original
matriz = np.array([[1,2],[3,4]])
submatriz = matriz[:,0].copy()
submatriz[0]= 200
print("matriz:\n", matriz)
print("submatriz:\n", submatriz)
```

¿Por qué es importante esto?

En ciencia de datos, es común transformar, limpiar y reestructurar datos. Si no distinguimos entre copiar y referenciar, podríamos:

• Corromper el conjunto de datos original por accidente.

• Propagar errores silenciosos en análisis.

• Obtener resultados inconsistentes en modelos o simulaciones.

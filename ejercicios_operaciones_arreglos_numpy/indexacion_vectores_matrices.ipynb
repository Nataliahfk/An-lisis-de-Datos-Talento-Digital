## Indexación y selección: selección de elementos de un arreglo.

En NumPy, indexar significa acceder a uno o varios elementos de un arreglo (ndarray) utilizando sus posiciones. Esta capacidad es esencial cuando necesitamos leer, modificar o extraer datos específicos para analizarlos o procesarlos. A diferencia de las listas tradicionales de Python, NumPy ofrece una variedad de formas para indexar datos que permiten seleccionar elementos de forma precisa y eficiente, tanto en arreglos unidimensionales (vectores) como en arreglos bidimensionales (matrices).

### Indexación en arreglos unidimensionales

Supongamos que tienes un arreglo con datos de velocidad registrados por un sensor:

```python
import numpy as np
```

```python
velocidades = np.array([30, 45, 55, 60, 48, 52])
```

## Acceso por posición

Podemos acceder a un valor individual indicando su índice (comienza en 0):

```python
print(velocidades[0]) # 30
print(velocidades[3]) # 60
```

## Índices negativos y rebanado

### Índices negativos

```python
# También puedes usar índices negativos para contar desde el final:
print(velocidades[-1]) # 52 (último valor)
print(velocidades[-3]) # 60 (tercer valor desde el final)
```

### Rebanado o slicing

```python
print(velocidades)
```

```python
# Si necesitas extraer una porción del arreglo:
segmento = velocidades[1:4] # considera de la posición 1 a la 3
print(segmento) # [45 55 60]
```

```python
segmento = velocidades [:3] # selecciona los tres primeros valores.
print(segmento)
```

```python
#selecciona desde la posición 3 hasta el final. Sin necesidad de conocer el tamaño del arreglo
segmento = velocidades [3:] 
print(segmento)
```

## Indexación en arreglos bidimensionales (matrices)

Cuando trabajamos con matrices (dos dimensiones), la indexación requiere dos coordenadas: fila y columna

```python
matriz = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(matriz)
```

### Acceso a elementos en matrices

```python
# Acceso a un elemento específico
print(matriz[1, 2]) # 60 (segunda fila(corresponde al 1, ya que comienza de 0 la fila y las columns), tercera columna)
```

```python
# Acceso a una fila completa
print(matriz[0]) # en este caso a la fila 0 o primera fila
print(matriz[0,:]) 
```

```python
# Acceso a una columna completa
print(matriz[:,1]) #todos los elemnetos de la fila correspondiente a la columna 1(segunda columna de todas las filas)
```

## Modificación de valores mediante indexación

```python
print(velocidades)
```

Podemos asignar nuevos valores usando índices:

```python
# en un vector
velocidades[2] = 50
print(velocidades)
```

```python
print(matriz)
```

```python
# en una matriz
matriz[1,1] = 88
print(matriz)
```

## Reemplazo de secciones y acceso avanzado

### Reemplazo de secciones completas

```python
print(velocidades)
```

Es posible reemplazar un segmento completo del arreglo con nuevos valores:

```python
velocidades[0:3] = [35, 46, 51]
print(velocidades)
```

❓ ¿Qué pasa si no coinciden?

Por ejemplo:

velocidades[0:3] = [35, 46]


Izquierda → 3 elementos

Derecha → 2 elementos

👉 No coinciden.

Resultado:

❌ NumPy lanza un error.

O al revés:

velocidades[0:3] = [35, 46, 51, 80]


Izquierda → 3 elementos

Derecha → 4 elementos

👉 Tampoco coincide.

❌ Error.

### Acceso con listas de índices

```python
print(velocidades)
```

Puedes seleccionar posiciones específicas usando una lista:

Esto se denomina indexación avanzada y es útil para aplicar filtros personalizados.

```python
print(velocidades)
```

```python
tamaño= len(velocidades)
tamaño
```

```python
posiciones = [0, 2, 4]
print(velocidades[posiciones])
```

## Indexación booleana

```python
# que imprima las velocidades mayores a 50
print(velocidades[velocidades > 50])
```

```python
print(matriz)
```

```python
print(matriz[matriz>30])
```

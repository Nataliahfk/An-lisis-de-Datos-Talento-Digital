## Ejercicio

### Objetivo de la actividad

Aplicar técnicas de indexación, selección condicional, operaciones entre arreglos y funciones estadísticas de NumPy, para analizar una matriz de datos meteorológicos simulados.

### Situación problema
Formas parte de un equipo técnico que analiza datos recogidos por estaciones meteorológicas automáticas distribuidas en distintas regiones del país. Cada estación registra información horaria sobre tres variables: temperatura (°C), humedad relativa (%) y velocidad del viento (km/h).

### Instrucciones

Tu jefatura te entrega una matriz de datos correspondiente a 8 horas consecutivas de medición de una estación y te solicita:

1. Extraer subconjuntos de interés según condiciones específicas. 
2. Comparar las mediciones con valores umbral definidos por la política interna de alertas.
3. Aplicar funciones estadísticas para obtener indicadores descriptivos.
4. Simular condiciones extremas para fines de prueba

crear una matriz simulada realista de una estación meteorológica con:

8 horas

3 variables por hora:

columna 0 → temperatura (°C)

columna 1 → humedad (%)

columna 2 → viento (km/h)

```python
import numpy as np
```

```python
datos = np.array([
    [18.5, 65, 12],
    [19.0, 63, 15],
    [20.3, 60, 18],
    [21.7, 58, 20],
    [22.1, 55, 25],
    [23.0, 52, 30],
    [24.5, 50, 35],
    [25.0, 48, 40]
])
```

```python
datos.shape # 8 horas 3 variables 
# Matriz de 8 filas 3 columnas temperatura (°C),humedad relativa (%) y velocidad del viento (km/h).
```

### 1️⃣ Extraer subconjuntos según condiciones

ej: 👉 horas donde la temperatura es mayor a 22°C

```python
# separar la columna temperatura
temperatura = datos[:,0] # toda las filas de la columna 0 que corresponde a temperaturas
print(temperatura)
```

```python
temperatura_mayor = datos[temperatura > 22]
print(temperatura_mayor)

''' 
si consulta temperaturas[temperaturas > 22] solo me revisara la columna aislada de temperatura filtras sobre el 
vector temperaturas y el resultado es un vector. En cambio en datos[temperatura > 22] usas la condición para 
filtrar filas de la matriz completa. En este caso dice Extraer subconjuntos de interés según condiciones específicas
👉 normalmente se espera: ✔ conservar todas las variables , por eso es mejor preguntar por datos[temperatura > 22].
'''
```

### 2️⃣ Comparar con valores umbral

Vamos a definir umbrales (típicos de alerta):

temperatura > 24 °C

humedad < 50 %

viento > 30 km/h

```python
alerta_temp = datos[:,0] > 24
alerta_humedad = datos[:,1] <= 50
alerta_viento = datos[:,2] > 30

print(alerta_temp)
print(alerta_humedad)
print(alerta_viento)
```

👉 Si queremos saber qué filas cumplen alguna alerta:

```python
filas_alerta = datos[alerta_temp | alerta_humedad | alerta_viento]
print(filas_alerta)
```

```python
valores_alerta_temp = datos[:,0][alerta_temp] # datos[:,0] > 24
valores_alerta_humedad = datos[:,1][alerta_humedad] # datos[:,1] < 50
valores_alerta_viento = datos[:,2][alerta_viento] # datos[:,2] > 30
print(valores_alerta_temp)
print(valores_alerta_humedad)
print(valores_alerta_viento)
```

```python
datos_alerta_temp = datos[alerta_temp]
datos_alerta_humedad = datos[alerta_humedad]
datos_alerta_viento = datos[alerta_viento]
print(datos_alerta_temp)
print(datos_alerta_humedad)
print(datos_alerta_viento)
```

### 3️⃣ Funciones estadísticas

El parámetro axis=0 indica que el cálculo se realiza por columnas, es decir, por cada variable meteorológica.

👉 promedio por variable:

```python
promedio = np.mean(datos,axis=0) # axis 0 es columna /  promedio temperatura, humedad,viento
print(promedio)
```

👉 máximo por variable:


```python
maximo = np.max(datos, axis=0)
print(maximo)
```

👉 mínimo por variable:

```python
minimo = np.min(datos, axis=0)
print(minimo)
```

👉 desviación estándar:

```python
desviscion = np.std(datos, axis= 0)
print(desviscion)
```

4️⃣ Simular condiciones extremas

Ahora copiamos la matriz para no dañar los datos originales.

```python
datos_extremos = datos.copy()
print(datos_extremos)
```

Se simulan condiciones extremas en las dos últimas horas:

- aumento de 5 °C en la temperatura

- aumento de 15 km/h en la velocidad del viento

```python
datos_extremos[-2:,0] += 5 # incremento en 5° temperatura las últimas 2 filas (24.5. a 29.5 y 25 a 30)
datos_extremos[-2:, 2] += 15 # incremento en 15 las últimas 2 filas (35. a 50 y 40 a 55)

print(datos_extremos)
```

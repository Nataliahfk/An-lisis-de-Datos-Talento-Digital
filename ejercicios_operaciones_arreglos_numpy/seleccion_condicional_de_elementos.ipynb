# Selección condicional de elementos de un arreglo.

En análisis de datos, es común tener que filtrar los valores que
cumplen una cierta condición: por ejemplo, seleccionar solo los
valores mayores a un umbral, encontrar los elementos que cumplan
múltiples criterios o extraer filas específicas de una matriz que
cumplan con cierta lógica. NumPy facilita este proceso mediante una
técnica llamada indexación booleana, que permite seleccionar
subconjuntos de datos en base a expresiones lógicas.

## Condiciones sobre arreglos unidimensionales

```python
import numpy as np
```

```python
# Supongamos que tienes un vector con registros de temperatura:
temperaturas = np.array([22.5, 25.0, 19.8, 27.3, 21.0])
```

## Aplicar una condición

Podemos generar un arreglo de valores True o False indicando si cada
elemento cumple la condición:

```python
# recorre el vector e indica True cuando la condicion es verdadera y False cuando es falsa y entrega un vector
condicion = temperaturas > 23
print(condicion)
```

### Usar la condición como índice

```python
# imprime las temperaturas que cumplen la condicion guardada en la variable condicion, en este caso > 23
print(temperaturas[condicion])
```

```python
# para seleccionar los que no cumplen con la condición // ~ simbolo para la negación
print(temperaturas[~condicion])
```

### Expresión compacta

Este enfoque nos permite filtrar automáticamente los datos.

Podemos hacer todo en una sola línea:

```python
# sobre el arreglo temperaturas realizo la condición temperaturas > 23
print(temperaturas[temperaturas > 23])
```

```python
print(temperaturas[temperaturas < 21])
print(temperaturas[temperaturas == 25])
print(temperaturas[temperaturas != 22.5])
```

## Múltiples condiciones (combinadas)

Podemos combinar condiciones con los operadores lógicos de NumPy:

• & significa AND lógico

• | significa OR lógico

• ~ significa NOT lógico

#### ¡Importante! Los paréntesis son obligatorios para que Python evalúe correctamente la operación.

Ejemplo: seleccionar temperaturas mayores a 20 y menores a 26

```python
print(temperaturas)
```

```python
condicion = (temperaturas > 20) & (temperaturas < 26)
print(condicion)
print(temperaturas[condicion])
```

```python
condicion = (temperaturas > 23) | (temperaturas == 22)
print(temperaturas[condicion])
```

```python
condicion = ~(temperaturas > 21)
print(temperaturas[condicion])
```

```python
print(temperaturas[~(temperaturas > 23)])
```

## Condiciones sobre matrices

También podemos aplicar condiciones a matrices completas. Supongamos:

```python
matriz = np.array([[11,20,33],[48,53,62],[74,88,95]])
print(matriz)
```

```python
# Obtener los elementos mayores a 20
print(matriz[matriz > 20])
```

Esto devuelve un vector con todos los elementos que cumplen la condición.

```python
print(matriz[~(matriz < 35)])
```

## Máscaras booleanas reutilizables

```python
# condicion, imprimir números pares de la matriz
pares = (matriz % 2 == 0)
print(matriz[pares])
```

```python
impares = (matriz % 2 == 0)
print(matriz[~(pares)])
```

## Filtrado combinado con asignación

Podemos usar condiciones no solo para seleccionar, sino también para modificar:

```python
print(temperaturas)
```

```python
temperaturas[temperaturas > 22] = 22
print(temperaturas)
```

Esto es útil en limpieza de datos, imputación de valores mínimos o normalización.
